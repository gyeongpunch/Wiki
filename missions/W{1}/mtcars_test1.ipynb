```python
total_cnt = len(mtcars.columns[1:])
for column in range(1, total_cnt + 1):
    plt.subplot(6, 2, column)
    plt.figure(figsize=(3, 2))  # 플롯 사이즈 설정 (선택사항)
    plt.hist(mtcars[mtcars.columns[column]], bins=20, color='blue', alpha=0.7)  # 히스토그램 그리기
    plt.xlabel(column)  # x축 레이블 설정
    plt.ylabel('Frequency')  # y축 레이블 설정
    plt.title(f'Histogram of {column}')  # 그래프 제목 설정
    plt.grid(True)  # 그리드 추가 (선택사항)
    plt.show()  # 그래프 출력
```
