```python
# 결과값을 바 그래프로 그리세요. 타이틀은 '# of Cars by Gear x Transmission'로, x축 레이블은 '(Gear, Auto Transmission)', 
# y축 레이블은 '# of Cars'로 만드세요.

new_x = []
new_value = []
for i in range(3, 5 + 1):
    for j in range(0, 1 + 1):
        new_x.append(str(i) + 'x' + str(j))
        new_value.append(mtcars[(mtcars['gear']==i) & (mtcars['am']==j)].car_name.nunique())
new_value
```

```python
plt.bar(new_x, new_value)
plt.xlabel('# of Cars by Gear x Transmission')
plt.ylabel('# of Cars')
```

```python
# 데이터셋에 포함된 모든 features의 히스토그램을 그리세요.
# 그래프 타이틀은 "Histograms of Variables"로 하세요.

total_cnt = len(mtcars.columns[1:])
for column in range(1, total_cnt + 1):
    
    plt.figure(figsize=(3, 2))
    plt.hist(mtcars[mtcars.columns[column]], bins=20, color='blue', alpha=0.7)
    plt.subplot(6, 2, column)
    plt.xlabel(column)
    plt.grid(True)
    plt.show()
```
